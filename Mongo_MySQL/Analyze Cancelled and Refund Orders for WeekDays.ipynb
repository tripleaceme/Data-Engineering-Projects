# Analyze Sales for Cancelled and Refund Orders for WeekDays

```python
# Open a connection to the database
with engine.connect() as connection:
    # Define a SQL query
    data = text("""
                    SELECT ors.customer_id, full_name, gender, age_group, state_code, 
                    ors.product_id, product_name, quantity, total_amount, order_date,
                    status, category, ratings, reviews, order_month, order_day, isWeekDay
                    
                    FROM customers cus 
                    JOIN orders ors ON cus.customer_id = ors.customer_id
                    JOIN products prs ON prs.product_id = ors.product_id
                    WHERE isWeekDay = 'Weekday' AND status IN ('Cancelled','Refund')
    
                """)
    
    # Execute the query and fetch the results into a pandas DataFrame
    all_data = pd.read_sql(data, connection)
```

```python
sales_data = all_data.copy()
```
